```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b:')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x))
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2,'y')
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'g:')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2,'y')
fig.tight_layout()
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'g:')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2,'y')
ax5.plot(y,np.cos(y))
ax6.plot(y,np.sin(y),'g:')
ax7.plot(y,np.tan(y),'r')
ax8.plot(y,np.cos(y)**2,'y')
fig.tight_layout()
```
