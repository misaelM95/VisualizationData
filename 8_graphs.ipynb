```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

HISTOGRAMA
-DISTRIBUCIÃ“N Y FREQUENCIA-

```python
plt.hist(data,bins=10,width=5,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

BOXPLOT -DISTRIBUCIÃ“N DE DATOS-

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

SCATTER -DISPERSIÃ“N- -SE UTILIZA PARA SABER SI TIENE CORRELACIÃ“N-

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
#market='p'->poligono
#market='x'->Nos arroja X's
#market='<'->Triangulos
#market='o'->Default
plt.scatter(x,y,s=area,c=colors,marker='*',alpha=0.5)
plt.show()
```
