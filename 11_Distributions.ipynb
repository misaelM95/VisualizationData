```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=True);
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='frequency');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='probability');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='percent');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='density');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='count',multiple='stack');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='count',multiple='dodge');
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,
             hue='sex',stat='count',multiple='fill');
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=True);
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',
            cumulative=False,shade=True,bw_adjust=1);
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count');
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist',multiple='stack');
```
