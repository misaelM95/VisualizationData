```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country =['INDIA','JAPAN','MEXICO','COLOMBIA','ALEMANIA']
population =[1000,800,1200,1300,1100]
```

```python
plt.bar(country,population,width=0.6,color=['red','green','yellow','blue','violet'])
plt.xticks(np.arange(5),('INDIA','JAPON','MEXICO','COLOMBIA','ALEMANIA'),rotation=45)
plt.title('PAISES CON MAYOR POPULARIDAD')
plt.xlabel('COUNTRYs')
plt.ylabel('POPULATION')
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
