```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
#Nos arroja los estilos predeterminados que existen
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'--')
ax.plot(x,x+3,'-')
ax.plot(x,x+5,':')
ax.plot(x,x+7,'o-')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color='green', alpha=0.99, linewidth=5,linestyle='--',marker='o')
ax.plot(x,x+3,linestyle='dashed')
ax.plot(x,x+5,linewidth=3)
ax.plot(x,x+7)
```
