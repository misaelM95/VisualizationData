```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=7,
            linecolor='black',vmin=0.5,vmax=1,cbar=False)
plt.show()
```
